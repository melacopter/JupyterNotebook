### My code is meant to display the temperature in my local area on a specific day ###

```python
import matplotlib.pyplot as plt
temp = [57, 56, 56, 56, 58, 67, 72, 74, 73, 68, 66, 64]
hour = ["12am", "2am", "4am", "6am", "8am", "10am", "12pm", "2pm", "4pm", "6pm", "8pm", "10pm"]
plt.plot(hour, temp)
plt.title('24hr Temperature Forecast for Thursday, November 18 in Norfolk, VA')
plt.xlabel('Hour')
plt.ylabel('Temperature')
plt.show()
```

The following code block sets the title, x-axis, and y-axis of the graph. The first line specifically determines the type of graph, in this case a line graph, and the data sets to be used in the graph displayed above:<br>
